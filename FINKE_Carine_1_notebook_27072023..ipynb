# jointure des tables

```python
## Ajout de la table product à transaction
products_transactions = transactions.merge(products, left_on='id_prod', right_on='id_prod', how='left')
products_transactions
```

```python
# Ajout de la table customers à product_transactions
products_transactions_cus = products_transactions.merge(customers, left_on='client_id', right_on='client_id', how='left')
products_transactions_cus
```
